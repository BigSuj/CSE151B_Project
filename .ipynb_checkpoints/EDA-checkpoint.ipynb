# Feature Engineering

```python
train_data['TRAVEL_TIME'] = train_data['POLYLINE'].apply(lambda x: max(len(x.split('[')) - 2, 0) * 15)
train_data
```

```python
timestamp = pd.to_datetime(train_data['TIMESTAMP'], unit = 's')
train_data["YEAR"] = timestamp.apply(lambda x: x.year)
train_data["MONTH"] = timestamp.apply(lambda x: x.month)
train_data["DAY"] = timestamp.apply(lambda x: x.day)
train_data["HOUR"] = timestamp.apply(lambda x: x.hour)
train_data["DAY_OF_WEEK"] = timestamp.apply(lambda x: x.dayofweek)
```

## Heatmap of GPS Coordinates

```python
# Converts the elements in the POLYLINE column from string to list

def create_matrix(raw):
    raw_0 = raw.split("],[")
    raw_1 = list(map(lambda x: x[2:] if x[0] == '[' else x, raw_0))
    raw_2 = list(map(lambda x: x[:-2] if x[-1] == ']' else x, raw_1))
    raw_3 = list(map(lambda x: x.split(','), raw_2))
    processed = np.vstack(raw_3).astype(float)
    return processed
```

```python
# Aggregates all GPS coordinates from the POLYLINE column into a numpy array

temp = train_data[(train_data['TRAVEL_TIME'] > 0) & (train_data['MISSING_DATA'] == False)]
to_plot = temp['POLYLINE'].apply(create_matrix)
all_data = np.vstack(to_plot)
```

```python
sample = all_data # [np.random.choice(all_data.shape[0], 70000000), :]
x = sample[:, 0]
y = sample[:, 1]

# Adjusted range to eliminate outliers

plt.hist2d(x, y, bins = 300, range=[[-9.6, -6.2], [36.5, 42.5]], norm = matplotlib.colors.LogNorm(), cmap = "hot")
plt.xlabel('Longitude')
plt.ylabel('Latitude')
```

### Temporal Variables with respect to CALL_TYPE

```python
var = 'DAY_OF_WEEK' # HOUR

pt = train_data.pivot_table(index = var, columns = 'CALL_TYPE', values = 'TRAVEL_TIME', aggfunc = 'count')

X = train_data[var].value_counts().sort_index().index
calltypeA = pt.iloc[:, 0]
calltypeB = pt.iloc[:, 1]
calltypeC = pt.iloc[:, 2]

X_axis = np.arange(len(X))
  
fig = plt.figure(100)
fig.set_figwidth(9)
fig.set_figheight(5)

plt.bar(X_axis - 0.2, calltypeA, 0.2, label = 'Call Type A')
plt.bar(X_axis, calltypeB, 0.2, label = 'Call Type B')
plt.bar(X_axis + 0.2, calltypeC, 0.2, label = 'Call Type C')

plt.xticks(X_axis, X)
plt.xlabel(var)
plt.ylabel("Number of Trips")
plt.title("CALL_TYPE vs. {0}".format(var) )
plt.legend()
plt.show()
```

## Analysis of Top 10 samples with highest training loss

```python
highest = pd.read_csv("./highest_training_IDs.csv")
highest
```

### Plotting the paths of the samples

```python
hd = train_data.join(highest, how = "right", lsuffix = '_t', rsuffix = '_h')[['TRIP_ID_t', 'POLYLINE']]
hd['POLYLINE'] = hd['POLYLINE'].apply(create_matrix)
```

```python
plt.figure()
for i in range(10):
    points = hd['POLYLINE'].iloc[i]
    plt.plot(points[:, 0], points[:, 1])
    plt.scatter(points[0:1, 0], points[0:1, 1], color = "red")
    plt.scatter(points[-1:, 0], points[-1:, 1], color = "blue")
plt.xlabel('Longitude')
plt.ylabel('Latitude')
```

```python
fig, axs = plt.subplots(5, 2)
fig.set_figheight(15)
for i in range(10):
    points = hd['POLYLINE'].iloc[i]
    axs[i // 2, i % 2].plot(points[:, 0], points[:, 1])
    axs[i // 2, i % 2].scatter(points[0:1, 0], points[0:1, 1], color = 'red')
    axs[i // 2, i % 2].scatter(points[-1:, 0], points[-1:, 1], color = 'blue')
    axs[i // 2, i % 2].tick_params(axis='both', which='major', labelsize=5.5)
plt.savefig("paths.png")
```
